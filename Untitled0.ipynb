```python
import os

# Function to add '.jpg' extension to all files in a folder
def add_extension_to_files(folder_path):
    if not os.path.exists(folder_path):
        print("Folder does not exist.")
        return

    if not os.path.isdir(folder_path):
        print("The provided path is not a directory.")
        return

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        # Construct the current file's full path
        file_path = os.path.join(folder_path, filename)
        # Check if it's a file
        if os.path.isfile(file_path):
            # Split the file name and extension
            name, ext = os.path.splitext(filename)
            # If the file doesn't already have the .jpg extension
            if ext != ".jpg":
                # Rename the file by appending .jpg
                new_file_path = os.path.join(folder_path, name + ".jpg")
                os.rename(file_path, new_file_path)
                print(f"Renamed {filename} to {name}.jpg")

# Example usage:
pathu = r'/content/drive/MyDrive/Saket Final Year Project Dataset/front'
add_extension_to_files(pathu)
```

```python
import os
import cv2
import dlib
import numpy as np
from facenet_pytorch import InceptionResnetV1, MTCNN
from tqdm import tqdm

# Initialize face detector and facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("/content/shape_predictor_68_face_landmarks.dat")

# Load FaceNet model
facenet_model = InceptionResnetV1(pretrained='vggface2').eval()

# Function to detect face and extract facial landmarks
def detect_face_landmarks(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    face_landmarks = []
    for face in faces:
        landmarks = predictor(gray, face)
        face_landmarks.append(landmarks)
    return faces, face_landmarks

# Function to align faces using facial landmarks
def align_face(image, landmarks):
    aligned_faces = []
    for landmark in landmarks:
        aligned_face = dlib.get_face_chip(image, landmark)
        aligned_faces.append(aligned_face)
    return aligned_faces

# Function to generate embeddings
def generate_embeddings(image, model):
    image = cv2.resize(image, (160, 160))
    image = np.transpose(image, (2, 0, 1))  # HWC to CHW format
    image = np.expand_dims(image, axis=0)
    with torch.no_grad():
        embeddings = model(torch.tensor(image).float()).numpy()
    return embeddings

# Path to folder containing images
folder_path = '/content/drive/MyDrive/Saket Final Year Project Dataset/front'

# Iterate over each image in the folder
for filename in tqdm(os.listdir(folder_path)):
    if filename.endswith(".jpg"):
        # Load image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Detect faces and extract facial landmarks
        faces, face_landmarks = detect_face_landmarks(image)

        # Align faces
        aligned_faces = align_face(image, face_landmarks)

        # Generate embeddings for aligned faces
        for aligned_face in aligned_faces:
            embeddings = generate_embeddings(aligned_face, facenet_model)
            # Save or process embeddings as needed
            print(embeddings.shape)  # Example: print shape of embeddings
```
